```python
pd.set_option('display.max_columns', 500)
pd.DataFrame()

### Use parse_dates to convert from generic object type to datetime64[ns, UTC] type.
df1 = pd.read_csv("data/kernel/purchase_orders.csv", parse_dates=["delivery_date", "created_date_time", "modified_date_time"])
print(df1.info())
```

```python
print(df1[['unit']].describe())

## Unique unit values are: ['KG', 'PUND']
print(df1['unit'].unique())
## Note: seems like unit_id 43.0 is associated with 'PUND' and unit id 40.0 with 'KG'

## Unique status values are: ['Closed', 'Deleted', 'Open']
print(df1['status'].unique())
## Note: It looks like the status_id to status mapping is {1: 'Open', 2: 'Closed', 3: 'Deleted'}

# (33171, 12)
print(df1.shape)

# The vast majority (~87%) of the orders are closed.
print("Value counts ahead:")
print(df1['status'].value_counts())
print(df1['status'].value_counts(normalize=True))
```
